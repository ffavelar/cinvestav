```python
import numpy as np
import matplotlib.pyplot as plt

data01 = np.loadtxt('DOPC/300k/rdf/rdf_DOPC-DOPC_500-600ns_av.xvg')
data02 = np.loadtxt('POPC/300k/rdf/rdf_POPC-POPC_500-600ns_av.xvg')
data03 = np.loadtxt('DLPC/300k/rdf/rdf_DLPC-DLPC_500-600ns_av.xvg')

l01 = 'DOPC'
l02 = 'POPC'
l03 = 'DLPC'

fig = plt.figure(figsize=(10,6))

plt.plot(data01[:,0], data01[:,1],label=l01,color="black")
plt.plot(data02[:,0], data02[:,1],label=l02,color="green")                    
plt.plot(data03[:,0], data03[:,1],label=l03,color="red")

plt.xlabel('$r (nm)$', fontsize=17)
plt.ylabel('$g_{2D}(r)$', fontsize=17)
plt.title('Systems RDF', fontsize=20)
plt.legend()
plt.xlim([0,3])
plt.ylim([0,1.5])

plt.show()
fig.savefig('rdfs.jpg',bbox_inches='tight')
```
